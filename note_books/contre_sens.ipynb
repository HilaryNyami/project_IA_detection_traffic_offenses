<h3 style="color: #4db8ff; margin-top: 0;">📍 A. Validation d'appartenance d'un vehicule : <code>if_point_in_polygone</code></h3>
<p style="margin: 0; padding-left: 10px;">
Cette première fonction permet de <b>limiter l'analyse aux zones critiques</b> de la chaussée. En utilisant l'algorithme <code>pointPolygonTest</code> d'OpenCV, nous vérifions si le centre d'un véhicule (détecté par YOLO) se trouve à l'intérieur de nos polygones de test (Région Gauche ou Droite).
</p>
<ul>
    <li><b>polygone</b> : C'est la liste des coordonnées (x, y) définie plus bas pour les régions(points des ROI). </li>
    <li><b>point</b> : C'est le centre de la bounding box envoyé par le tracker de YOLO. </li>
    <li><b>False</b> : permet de demander si le point est dedans ou hors (pas la distance exacte) du polygone </li>
</ul>
</div>

```python
def if_point_in_polygone(polygone, point):
    return cv2.pointPolygonTest(polygone, point, False)>=0
```

    <h3 style="color: #4db8ff; margin-top: 0;">📏 Determiner la direction d'un vehicule : <code>get_direction_vector</code>
    </h3><p style="margin: 0; padding-left: 10px;">Pour décider si un véhicule est en contre-sens, il ne suffit pas de connaître sa position instantanée, il faut comprendre sa <b>trajectoire</b>. C'est ici qu'intervient la notion de mémoire temporelle :Cette fonction stocke les positions successives de chaque véhicule dans un <b>historique de 10 frames</b>. En calculant la différence entre le tout premier point capturé et le dernier (le <b>$d_x$</b>), nous obtenons un vecteur qui nous indique clairement la direction réelle du mouvement, tout en filtrant les petites oscillations ou les erreurs de détection passagères dues aux reflets de nuit.</p>
</div>

```python
#Calcule le vecteur direction d'un véhicule
#Retourne (dx, dy) : déplacement moyen sur les dernières frames
def get_direction_vector(vehicle_id, current_center, max_history=10):
    # Ajouter la position actuelle
    vehicle_history[vehicle_id].append(current_center)
    
    # Garder seulement les dernières positions
    if len(vehicle_history[vehicle_id]) > max_history:
        vehicle_history[vehicle_id].pop(0)
    
    # Si on n'a pas assez de points, retourner (0, 0)
    if len(vehicle_history[vehicle_id]) < 2:
        return (0, 0)
    
    # Calculer le déplacement moyen entre le premier et dernier point
    first_point = vehicle_history[vehicle_id][0]
    last_point = vehicle_history[vehicle_id][-1]
    
    dx = last_point[0] - first_point[0]
    dy = last_point[1] - first_point[1]
    
    return (dx, dy)
```

<h4 style="color: #2563EB">Fonction qui determine si un vehicule monte ou descend (de quel cote de la route est le vehicule)</h4>

```python
#Détermine la direction
#dx, dy : vecteur directeur (peut être négatif ou positif)
def sens_direction_object(dx, dy):
    if dx>0 : #le vehicule est du cote gauche donc sa valeur sur l'axe des abscisses augmente: il descend
        return 1
    elif dx<0: #le vehicule est du cote droit donc sa valeur sur l'axe des abscisses decroit: il monte
        return 2
    else:
        return 0
```

<h4 style="color: #2563EB">Fonction qui verifie capture la plaque d'immatriculation et la lit</h4>

```python
def extraire_et_lire_plaque(frame, bbox):
    x, y, w, h = bbox
    
    # Zone approximative de la plaque (partie arrière basse du véhicule)
    plaque_hauteur = int((h - y) * 0.25)  # 25% du bas du véhicule
    plaque_y1 = y
    plaque_y2 = h + plaque_hauteur
    
    # Ajuster les limites
    plaque_y1 = max(0, plaque_y1)
    plaque_y2 = min(frame.shape[0], plaque_y2)
    
    # Extraire la zone de la plaque
    zone_plaque = frame[plaque_y1.int():plaque_y2.int(), x.int():w.int()]
    
    if zone_plaque.size == 0:
        return None, None
    
    # Améliorer le contraste pour une meilleure lecture OCR
    zone_gris = cv2.cvtColor(zone_plaque, cv2.COLOR_BGR2GRAY)
    _, zone_amelioree = cv2.threshold(zone_gris, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # Essayer de lire avec EasyOCR
    try:
        resultats = reader.readtext(zone_amelioree)
        if resultats:
            # Prendre la détection avec la plus haute confiance
            meilleur_resultat = max(resultats, key=lambda x: x[2])
            texte_plaque = meilleur_resultat[1].replace(' ', '').upper()
            confiance = meilleur_resultat[2]
            
            # Filtrer les résultats peu probables
            if confiance > 0.6 and len(texte_plaque) >= 4:
                return texte_plaque, zone_amelioree
    except:
        pass
    
    return None, None
```

```python
if not video.isOpened():
    print('Erreur lors de l\'ouverture de la video')
else:
    print('Affichage de la video')
    while True:
        state, frame = video.read()
        cv2.polylines(frame, [region_gauche], True, (255, 0, 0), 2) # Bleu, épaisseur 2
        cv2.polylines(frame, [region_droite], True, (255, 0, 0), 2) # Bleu, épaisseur 2
         # Ajouter des labels aux régions
        cv2.putText(frame, "Region Gauche", (region_gauche[0][0], region_gauche[0][1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
        cv2.putText(frame, "Region Droite", (region_droite[0][0], region_droite[0][1] - 10),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

        
        if state:
            #Créer le VideoWrite

            #====================== tracking ===================
            # Run YOLO11 tracking on the frame, persisting tracks between frames
            result = model.track(frame, persist=True, verbose=False)[0]
            #save_video = model.predict(source=video_path, save=True)
            # Get the boxes and track IDs
            if result.boxes and result.boxes.is_track:
                boxes = result.boxes.xywh.cpu()
                track_ids = result.boxes.id.int().cpu().tolist()

    
                # Plot the tracks
                for box, track_id in zip(boxes, track_ids):
                    x, y, w, h = box
                    track = track_history[track_id]
                    track.append((float(x), float(y)))  # x, y center point
                    #print(track_id)
                    if len(track) > 10:  # retain 10 tracks for 10 frames
                        track.pop(0)

                    center_x = (x + w) // 2
                    center_y = (y + h) // 2
                    current_center = (float(x), float(y))
                    

                    # Si on n'a pas assez de points, retourner (0, 0)
                    if len(track_history[track_id]) < 4:
                        dx,dy = (0, 0)

                    else:
                    
                    # Calculer le vecteur directeur entre le premier et dernier point
                        first_point = track_history[track_id][0]
                        last_point = track_history[track_id][-1]
                        
                        dx = last_point[0] - first_point[0]
                        dy = last_point[1] - first_point[1]
                        

                            # Déterminer la direction
                    direction_vehicule = sens_direction_object(dx, dy)
                   
                    in_left_region = if_point_in_polygone(region_gauche, current_center)
                    in_right_region = if_point_in_polygone(region_droite, current_center)

                    contre_sens = False
                            # Si dans région gauche
                    if direction_vehicule != 0:
                    
                        if in_left_region:
                                    # En région gauche, on s'attend à ce que les véhicules aillent vers la droite (dx > 0)
                            if direction_vehicule != 1:  
                                contre_sens = True                            
                                # Si dans région droite
                        elif in_right_region:
                                    # En région droite, on s'attend à ce que les véhicules aillent vers la gauche (dx < 0)
                            if direction_vehicule == 1: 
                                contre_sens = True
                                
                           
                    if contre_sens:
                        # Dessiner
                        color = (0,0,255)
                        cv2.rectangle(frame, (int(x - w/2), int(y - h/2)), (int(x + w/2), int(y + h/2)), color, 3)
                        cv2.putText(frame, "CONTRE-SENS!", (int(x), int(y) - 40), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 3)
                        texte_plaque, zone = extraire_et_lire_plaque(frame, box)
                        cv2.putText(frame, texte_plaque, (5, 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (150, 150, 150), 3)
                    else:
                        color = (0, 255, 0)  # Vert pour normal
                        cv2.rectangle(frame, (int(x - w/2), int(y - h/2)), (int(x + w/2), int(y + h/2)), color, 2)
                        
                    video_out.write(frame)  # Sauvegarde la frame traitée
                    


            # Afficher la frame
            cv2.imshow('Detection des CONTRE-SENS', frame)            
            if cv2.waitKey(25) & 0xFF == ord('q'):
                break
        else:
            print('Fin de la video')
            break
    
    video.release()
    video_out.release()
    cv2.destroyAllWindows()
```

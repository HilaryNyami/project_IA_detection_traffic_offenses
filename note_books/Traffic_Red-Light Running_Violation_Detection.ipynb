    <p style="margin-top: 0; padding-left: 10px;">
        Dans cette section, l'objectif principal est de détecter la couleur du feu de circulation en temps réel à partir de la trame vidéo. Voici la décomposition des étapes du traitement d'image :
    </p>

<ol style="padding-left: 30px;">
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;"> Définition de la Région d'Intérêt (ROI) :</b> Nous commençons par définir une zone spécifique de l'image. Cette <b>ROI</b> est isolée pour concentrer l'analyse là où le feu de circulation est censé se trouver, réduisant ainsi la charge de calcul.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;"> Conversion en Espace HSV :</b> La ROI est convertie de l'espace couleur RGB à l'espace **HSV** (Teinte, Saturation, Valeur).  Le modèle HSV est privilégié en vision par ordinateur car il sépare la couleur pure (Teinte) de la luminosité, simplifiant la détection colorimétrique.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;"> Définition des Plages de Couleurs :</b> Nous définissons les plages de valeurs de Teinte (Hue) correspondant spécifiquement au **Rouge** et au **Jaune**. Ces seuils sont cruciaux pour l'identification du statut du feu.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;"> Création des Masques de Couleurs :</b> En utilisant les plages définies, des masques binaires sont générés. Ces masques mettent en évidence les régions de la ROI qui contiennent les couleurs Rouge et Jaune.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;"> Reconnaissance du Statut :</b> Nous examinons les masques.
            <ul>
                <li>Si le masque rouge contient des pixels non nuls : le feu est **Rouge** (Stop).</li>
                <li>Sinon, si le masque jaune contient des pixels non nuls : le feu est **Jaune** (Prudence).</li>
                <li>Si ni rouge ni jaune n'est détecté : le feu est considéré **Vert** (Passer).</li>
            </ul>
        </li>
        <li style="margin-bottom: 0;">
            <b style="color: #4db8ff;"> Affichage de l'Information :</b> Un message textuel indiquant le statut détecté (Stop/Prudence/Passer) est superposé sur la trame principale pour un retour visuel immédiat.
        </li>
    </ol>
    
<hr style="border-color: #1e2d38;">

<p style="margin-bottom: 0; padding-left: 10px;">
        La fonction <code style="background-color: #1e2d38; padding: 3px 6px; border-radius: 5px; color: #4db8ff;">detect_traffic_light_color</code> nous permet ainsi d'obtenir l'information vitale pour le traitement ultérieur et la détection d'infractions.
    </p>
</div>

```python
def detect_traffic_light_color(image, rect):
    # Extract rectangle dimensions
    x, y, w, h = rect
    # Extract region of interest (ROI) from the image based on the rectangle
    roi = image[y:y+h, x:x+w]
    
    # Convert ROI to HSV color space
    hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

    # Define HSV range for red color
    red_lower = np.array([0, 120, 70])
    red_upper = np.array([10, 255, 255])
    
    # Define HSV range for yellow color
    yellow_lower = np.array([20, 100, 100])
    yellow_upper = np.array([30, 255, 255])

    # Create binary masks for detecting red and yellow in the ROI
    red_mask = cv2.inRange(hsv, red_lower, red_upper)
    yellow_mask = cv2.inRange(hsv, yellow_lower, yellow_upper)
    
    # Font details for overlaying text on the image
    font = cv2.FONT_HERSHEY_TRIPLEX
    font_scale = 1  
    font_thickness = 2  
    
    # Check which color is present based on the masks
    if cv2.countNonZero(red_mask) > 0:
        text_color = (0, 0, 255)
        message = "Signal Status: Stop"
        color = 'red'
    elif cv2.countNonZero(yellow_mask) > 0:
        text_color = (0, 255, 255)
        message = "Signal Status: Attention"
        color = 'yellow'
    else:
        text_color = (0, 255, 0)
        message = "Signal Status: Go"
        color = 'green'

    cv2.putText(image, 30*'-', (10, 30), font, font_scale, (255,255,255), font_thickness, cv2.LINE_AA)
    # Overlay the detected traffic light status on the main image
    cv2.putText(image, message, (15, 70), font, font_scale+0.5, text_color, font_thickness+1, cv2.LINE_AA)
    # Add a separator line
    cv2.putText(image, 30*'-', (10, 115), font, font_scale, (255,255,255), font_thickness, cv2.LINE_AA)
    
    # Return the modified image and detected color
    return image, color
```

        Dans cette partie, nous définissons la classe `LineDetector`. Son objectif est de détecter et de stabiliser la ligne d'arrêt sur les trames vidéo, puis de la colorer dynamiquement en fonction du statut du feu de circulation (rouge, jaune ou vert).
    </p>
    
<h4 style="color: #4db8ff; margin-top: 20px; border-bottom: 1px solid #1e2d38; padding-bottom: 5px;">
        🔍 Stabilité et Conception Orientée Objet
    </h4>
    <p style="padding-left: 10px;">
        En raison de l'instabilité naturelle des lignes dans les flux vidéo (phénomène de *jittering*), il est essentiel de maintenir une représentation stable. Nous utilisons des **<b style="color: #4db8ff;">files d'attente</b>** (`deque`) pour stocker les positions antérieures de la ligne, ce qui permet de lisser les changements brusques et d'assurer une cohérence visuelle.
        
<br>
        L'implémentation sous forme de **classe** est cruciale car elle garantit la persistance de ces files d'attente entre les appels de trames, empêchant leur réinitialisation et assurant la stabilité de la détection.
    </p>

<h4 style="color: #4db8ff; margin-top: 20px; border-bottom: 1px solid #1e2d38; padding-bottom: 5px;">
        ⚙️ Opérations Principales de la Classe
    </h4>
    <ol style="padding-left: 30px;">
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">1. Initialisation :</b> Création des files d'attente (`y_start_queue` et `y_end_queue`) pour le lissage des coordonnées Y des points de départ et de fin de la ligne.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">2. Mappage des Couleurs :</b> Une fonction interne (`get_color_code`) mappe les noms de couleurs (rouge, vert, jaune) à leurs équivalents BGR pour la coloration dynamique de la ligne.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">3. Équations de Ligne :</b> Formulation de trois équations de ligne pour délimiter la zone d'intérêt de la ligne d'arrêt.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">4. Création des Masques :</b> Génération de masques progressifs basés sur les équations pour isoler la ligne d'arrêt tout en minimisant les distractions.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">5. Prétraitement d'Image :</b> Application de transformations classiques (gris, flou, égalisation d'histogramme, détection de contours) sur le masque raffiné.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">6. Détection de Ligne :</b> Utilisation de la **<b style="color: #4db8ff;">Transformée de Hough pour les lignes</b>** (Hough Line Transform) pour identifier la ligne d'arrêt. Les paramètres de la ligne détectée sont ensuite stockés dans nos files d'attente pour stabilisation. 
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">7. Dessin et Coloration :</b> La ligne stabilisée est dessinée sur la trame originale, sa couleur étant ajustée en temps réel selon le statut du feu de circulation.
        </li>
        <li style="margin-bottom: 0;">
            <b style="color: #4db8ff;">8. Masque Final de Zone d'Infraction :</b> Un masque est généré pour noircir tous les pixels **au-dessus** de la ligne d'arrêt détectée. Ce masque est l'élément clé pour l'étape suivante, car il permet d'isoler uniquement les plaques d'immatriculation des véhicules qui ont **franchi** la ligne rouge.
        </li>
    </ol>
    
<hr style="border-color: #1e2d38;">

<p style="margin-bottom: 0; padding-left: 10px;">
        La classe <code style="background-color: #1e2d38; padding: 3px 6px; border-radius: 5px; color: #4db8ff;">LineDetector</code> est ainsi la pierre angulaire de la surveillance, assurant une détection stable et localisant précisément la zone d'infraction.
    </p>
</div>

```python
class LineDetector:
    def __init__(self, num_frames_avg=10):
        # Initialize two deque queues to hold y-coordinate values across frames
        self.y_start_queue = deque(maxlen=num_frames_avg)
        self.y_end_queue = deque(maxlen=num_frames_avg)

    
    def detect_white_line(self, frame, color, 
                          slope1=0.03, intercept1=920, slope2=0.03, intercept2=770, slope3=-0.8, intercept3=2420):
        
        # Function to map color names to BGR values
        def get_color_code(color_name):
            color_codes = {
                'red': (0, 0, 255),
                'green': (0, 255, 0),
                'yellow': (0, 255, 255)
                 }
            return color_codes.get(color_name.lower())

        frame_org = frame.copy()
        
        # Line equations for defining region of interest (ROI)
        def line1(x): return slope1 * x + intercept1
        def line2(x): return slope2 * x + intercept2
        def line3(x): return slope3 * x + intercept3

        height, width, _ = frame.shape
        
        # Create a mask to spotlight the line's desired area
        mask1 = frame.copy()
        # Set pixels below the first line to black in mask1
        for x in range(width):
            y_line = line1(x)
            mask1[int(y_line):, x] = 0

        mask2 = mask1.copy()
        # Set pixels above the second line to black in mask2
        for x in range(width):
            y_line = line2(x)
            mask2[:int(y_line), x] = 0

        mask3 = mask2.copy()
        # Set pixels to the left of the third line to black in mask3 (final mask)
        for y in range(height):
            x_line = line3(y)
            mask3[y, :int(x_line)] = 0

        # Convert the mask to grayscale
        gray = cv2.cvtColor(mask3, cv2.COLOR_BGR2GRAY)

        # Apply a Gaussian filter to the ROI
        blurred_gray = cv2.GaussianBlur(gray, (7, 7), 0)

        # Apply CLAHE to equalize the histogram
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        gray_clahe = clahe.apply(blurred_gray)

        # Perform edge detection
        edges = cv2.Canny(gray, 30, 100)

        # Perform a dilation and erosion to close gaps in between object edges
        dilated_edges = cv2.dilate(edges, None, iterations=1)
        edges = cv2.erode(dilated_edges, None, iterations=1)

        # Perform Hough Line Transform
        lines = cv2.HoughLinesP(edges, 1, np.pi/180, 100, minLineLength=160, maxLineGap=5)

        # Calculate x coordinates for the start and end of the line
        x_start = 0
        x_end = width - 1

        if lines is not None:
            for line in lines:
                x1, y1, x2, y2 = line[0]

                # Calculate line parameters
                slope = (y2 - y1) / (x2 - x1 + np.finfo(float).eps)  # Add a small number to avoid division by zero
                intercept = y1 - slope * x1

                # Calculate corresponding y coordinates
                y_start = int(slope * x_start + intercept)
                y_end = int(slope * x_end + intercept)

                # Add the y_start and y_end values to the queues
                self.y_start_queue.append(y_start)
                self.y_end_queue.append(y_end)

        # Compute the average y_start and y_end values
        avg_y_start = int(sum(self.y_start_queue) / len(self.y_start_queue)) if self.y_start_queue else 0
        avg_y_end = int(sum(self.y_end_queue) / len(self.y_end_queue)) if self.y_end_queue else 0

        
        # Draw the line
        line_start_ratio=0.32
        x_start_adj = x_start + int(line_start_ratio * (x_end - x_start))  # Adjusted x_start
        avg_y_start_adj = avg_y_start + int(line_start_ratio * (avg_y_end - avg_y_start))  # Adjusted avg_y_start

        # Create a mask with the same size as the frame and all zeros (black)
        mask = np.zeros_like(frame)

        # Draw the line on the mask
        cv2.line(mask, (x_start_adj, avg_y_start_adj), (x_end, avg_y_end), (255, 255, 255), 4)

        # Determine which color channel(s) to change based on the color argument
        color_code = get_color_code(color)
        if color_code == (0, 255, 0):  # Green
            channel_indices = [1]
        elif color_code == (0, 0, 255):  # Red
            channel_indices = [2]
        elif color_code == (0, 255, 255):  # Yellow
            # Yellow in BGR is a combination of green and red channels.
            # Here we modify both green and red channels.
            channel_indices = [1, 2]
        else:
            raise ValueError('Unsupported color')

        # Change the specified color channels of the frame where the mask is white
        for channel_index in channel_indices:
            frame[mask[:,:,channel_index] == 255, channel_index] = 255
                
                
        # Calculate slope and intercept for the average green line
        slope_avg = (avg_y_end - avg_y_start) / (x_end - x_start + np.finfo(float).eps)
        intercept_avg = avg_y_start - slope_avg * x_start

        # Create a mask with the pixels above the green line set to black
        mask_line = np.copy(frame_org)
        for x in range(width):
            y_line = slope_avg * x + intercept_avg - 35
            mask_line[:int(y_line), x] = 0  # set pixels above the line to black

        return frame, mask_line
```

Son rôle est de détecter et d'extraire les plaques d'immatriculation à partir de la trame vidéo, en se concentrant spécifiquement sur la zone d'infraction. Voici le processus détaillé :
    </p>

<ol style="padding-left: 30px;">
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">1. Prétraitement d'Image :</b> La trame subit un prétraitement essentiel :
            <ul>
                <li>Conversion en niveaux de gris (pour les valeurs d'intensité monochromes).</li>
                <li>Application de **<b style="color: #4db8ff;">CLAHE</b>** (Contrast Limited Adaptive Histogram Equalization) pour améliorer le contraste local.</li>
                <li>Réduction du bruit par érosion, préparant l'image pour une détection plus précise.</li>
            </ul>
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">2. Isolement et Recadrage de la ROI :</b>
            Le code utilise le masque de l'étape précédente (où la zone au-dessus de la ligne d'arrêt est noircie) pour identifier les pixels de la route *en infraction*. Un cadre englobant (bounding box) est calculé pour cette zone.  L'image est ensuite recadrée, en excluant une partie de la voie opposée. Cette **ROI recadrée et bien définie** permet au classifieur en cascade de travailler de manière plus efficace.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">3. Classification par Cascade de Haar :</b>
            L'image recadrée en niveaux de gris est soumise à un **<b style="color: #4db8ff;">classifieur en cascade de Haar</b>**. Cette méthode classique (non basée sur le Deep Learning) est utilisée pour sa capacité à reconnaître des motifs. Pré-entraîné pour les plaques d'immatriculation, il renvoie les emplacements potentiels sous forme de rectangles. 
        </li>
        <li style="margin-bottom: 0;">
            <b style="color: #4db8ff;">4. Extraction et Sortie Finale :</b>
            Les plaques détectées sont :
            <ul>
                <li>Marquées sur la trame originale par des rectangles.</li>
                <li>Recadrées de la zone en niveaux de gris et stockées dans une liste.</li>
            </ul>
            La fonction retourne la trame originale avec les cadres et la liste des plaques d'immatriculation potentielles.
        </li>
    </ol>
    
 <hr style="border-color: #1e2d38;">

<p style="margin-bottom: 0; padding-left: 10px;">
        La fonction <code style="background-color: #1e2d38; padding: 3px 6px; border-radius: 5px; color: #4db8ff;">extract_license_plate</code> est une solution optimisée et structurée qui exploite les techniques de vision par ordinateur classiques pour extraire efficacement l'information critique.
    </p>
</div>

```python
def extract_license_plate(frame, mask_line):    
    # Convert the image to grayscale (Haar cascades are typically trained on grayscale images)
    gray = cv2.cvtColor(mask_line, cv2.COLOR_BGR2GRAY)
    
    # Apply CLAHE to equalize the histogram
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    gray = clahe.apply(gray)
    
    # Erode the image using a 2x2 kernel to remove noise
    kernel = np.ones((2, 2), np.uint8)
    gray = cv2.erode(gray, kernel, iterations=1)

    # Find the bounding box of non-black pixels
    non_black_points = cv2.findNonZero(gray)
    x, y, w, h = cv2.boundingRect(non_black_points)

    # Calculate the new width of the bounding box, excluding 30% on the right side
    w = int(w * 0.7)

    # Crop the image to the bounding box
    cropped_gray = gray[y:y+h, x:x+w]

    # Detect license plates in the image (this returns a list of rectangles)
    license_plates = license_plate_cascade.detectMultiScale(cropped_gray, scaleFactor=1.07, minNeighbors=15, minSize=(20, 20))

    # List to hold cropped license plate images
    license_plate_images = []

    # Loop over the license plates
    for (x_plate, y_plate, w_plate, h_plate) in license_plates:
        # Draw a rectangle around the license plate in the original frame (here you need the original coordinates)
        cv2.rectangle(frame, (x_plate + x, y_plate + y), (x_plate + x + w_plate, y_plate + y + h_plate), (0, 255, 0), 3)
    
        # Crop the license plate and append it to the list (here x_plate and y_plate are relative to cropped_gray)
        license_plate_image = cropped_gray[y_plate:y_plate+h_plate, x_plate:x_plate+w_plate]
        license_plate_images.append(license_plate_image)

    return frame, license_plate_images
```

Elle est conçue pour reconnaître et extraire l'information textuelle (caractères et chiffres) des images de plaques d'immatriculation extraites, en utilisant l'outil d'**OCR PyTesseract**.

 <h4 style="color: #4db8ff; margin-top: 20px; border-bottom: 1px solid #1e2d38; padding-bottom: 5px;">
        ⚙️ Flux Opérationnel de l'OCR
    </h4>
    <ol style="padding-left: 30px;">
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">1. Seuillage de l'Image (Thresholding) :</b> L'image de la plaque est transformée en un format binaire (noir et blanc).  Cette opération est essentielle pour créer un contraste maximal, garantissant que les caractères sont distincts de l'arrière-plan avant l'étape de reconnaissance.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">2. Conversion de Format :</b> L'image, initialement au format OpenCV, est convertie au format **PIL (Pillow Image)**. Cette étape assure la compatibilité avec le moteur d'OCR **PyTesseract**.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">3. Extraction de Texte par OCR Classique :</b> L'outil PyTesseract est appliqué à l'image convertie. Une configuration spéciale (`--psm 6`) est utilisée. Ce paramètre (`PSM 6`) est optimisé pour la reconnaissance de **texte uniforme sur une seule ligne**, ce qui est idéal pour les plaques d'immatriculation.
        </li>
        <li style="margin-bottom: 0;">
            <b style="color: #4db8ff;">4. Nettoyage du Texte :</b> Après l'extraction, tous les espaces blancs ou caractères superflus potentiellement ajoutés au début ou à la fin du texte reconnu sont supprimés (`strip()`) pour garantir que la plaque retournée est propre et compacte.
        </li>
    </ol>
    
<hr style="border-color: #1e2d38;">

<p style="margin-bottom: 0; padding-left: 10px;">
        Grâce à la fonction <code style="background-color: #1e2d38; padding: 3px 6px; border-radius: 5px; color: #4db8ff;">apply_ocr_to_image</code>, nous exploitons la puissance des techniques d'apprentissage automatique traditionnelles pour l'OCR, permettant d'obtenir l'information textuelle nécessaire à la verbalisation.
    </p>
</div>

```python
def apply_ocr_to_image(license_plate_image):    
    # Threshold the image
    _, img = cv2.threshold(license_plate_image, 120, 255, cv2.THRESH_BINARY)

    # Convert OpenCV image format to PIL Image format for pytesseract
    pil_img = Image.fromarray(img)

    # Use pytesseract to extract text from the image
    full_text = pytesseract.image_to_string(pil_img, config='--psm 6')

    return full_text.strip()  # Removing any extra white spaces from the ends
```

```python
def create_database_and_table(host, user, password, database):
    try:
        # Create a connection
        connection = mysql.connector.connect(
            host = host,
            user = user,
            password = password
        )
        
        if connection.is_connected():
            # Create a new database cursor
            cursor = connection.cursor()

            # Create a new database using the provided name
            cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database}")
            print(f"Database {database} created successfully!")

            # Use the newly created database
            cursor.execute(f"USE {database}")

            # Create a new table
            cursor.execute("""
                CREATE TABLE IF NOT EXISTS license_plates (
                    id INT AUTO_INCREMENT PRIMARY KEY,
                    plate_number VARCHAR(255) NOT NULL UNIQUE,
                    violation_count INT DEFAULT 1
                )
            """)
            print("Table créée avec succès!")

            cursor.close()

    except Error as e:
        print("Error while connecting to MySQL", e)

    finally:
        if connection.is_connected():
            connection.close()
```

    </p>

<p style="padding-left: 10px;">
        Cette fonction est la clé de la persistance des données. Elle gère l'enregistrement d'une infraction avec la logique suivante :
    </p>

<ul style="padding-left: 30px;">
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">Connexion :</b> Se connecte à la base de données MySQL déjà configurée.
        </li>
        <li style="margin-bottom: 12px;">
            <b style="color: #4db8ff;">Vérification d'Existence :</b> Vérifie si la plaque d'immatriculation concernée par l'infraction est **déjà présente** dans la table.
        </li>
        <li style="margin-bottom: 0;">
            <b style="color: #4db8ff;">Logique de Mise à Jour :</b> 
            <ul>
                <li>Si la plaque existe : Son **compteur d'infractions est incrémenté** (`violation count is incremented`).</li>
                <li>Si la plaque n'existe pas : Un **nouvel enregistrement** est ajouté avec un compteur par défaut de 1.</li>
            </ul>
        </li>
    </ul>

<hr style="border-color: #1e2d38;">

<p style="margin-bottom: 0; padding-left: 10px;">
        Cette approche garantit que l'historique des infractions par véhicule est maintenu de manière précise et efficace.
    </p>
</div>

```python
def update_database_with_violation(plate_number, host, user, password, database):
    try:
        connection = mysql.connector.connect(
            host = host,
            user = user,
            password = password,
            database = database
        )
        
        if connection.is_connected():
            cursor = connection.cursor()

            # Check if the license plate already exists in the table
            cursor.execute(f"SELECT violation_count FROM license_plates WHERE plate_number='{plate_number}'")
            result = cursor.fetchone()
            
            if result:
                # Increment violation_count by 1 if plate_number already exists
                cursor.execute(f"UPDATE license_plates SET violation_count=violation_count+1 WHERE plate_number='{plate_number}'")
            else:
                # Insert a new record if plate_number doesn't exist
                cursor.execute(f"INSERT INTO license_plates (plate_number) VALUES ('{plate_number}')")
            
            connection.commit()
            cursor.close()

    except Error as e:
        print("Error while connecting to MySQL", e)

    finally:
        if connection.is_connected():
            connection.close()
```

```python
def main():
    # Ensure the database and table exist
    create_database_and_table(DB_HOST, DB_USER, DB_PASSWORD, DB_NAME)

    # Clear the license plates from the previous run. (Comment out this line if desired!)
    clear_license_plates(DB_HOST, DB_USER, DB_PASSWORD, DB_NAME)

    # Open the video file
    vid = cv2.VideoCapture('traffic_video.mp4')
    
    # Create detector object
    detector = LineDetector()

    # Loop through each frame in the video
    while True:
        # Read frame
        ret, frame = vid.read()
        
        # Break if frame is not returned
        if not ret:
            break

        # Assuming rect is the rectangle where the traffic light is located
        rect = (1700, 40, 100, 250) 
        
        # Detect traffic light color
        frame, color = detect_traffic_light_color(frame, rect)
        
        # Detect white line
        frame, mask_line = detector.detect_white_line(frame, color)
        
        # Process the frame if the light is red
        if color == 'red':
            # Extract license plate
            frame, license_plate_images = extract_license_plate(frame, mask_line)
            
            # Process each detected license plate
            for license_plate_image in license_plate_images:
                # Apply OCR to the license plate image
                text = apply_ocr_to_image(license_plate_image)
                
                # Add the detected license plate to the list if it matches the pattern and is not already in the list
                if text is not None and re.match("^[A-Z]{2}\s[0-9]{3,4}$", text) and text not in penalized_texts:
                    penalized_texts.append(text)
                    print(f"\nPlaque d'immatriculation verbalisée: {text}")

                    # Plot the license plate image
                    plt.figure()
                    plt.imshow(license_plate_image, cmap='gray')
                    plt.axis('off')
                    plt.show()
                    
                    # Update the database with the license plate violation
                    update_database_with_violation(text, DB_HOST, DB_USER, DB_PASSWORD, DB_NAME)
        
        # Draw the penalized text onto the frame if there is any
        # if penalized_texts:
        #     draw_penalized_text(frame)
            
        # Permet de redimensionner la fenêtre
        cv2.namedWindow('frame', cv2.WINDOW_NORMAL)  
        cv2.resizeWindow('frame', 800, 600)

        # Display the frame
        cv2.imshow('frame', frame)

        # Break if ESC key is pressed
        if cv2.waitKey(1) == 27:
            break

    # Release the video
    vid.release()
    
    # Close all OpenCV windows
    cv2.destroyAllWindows()

    # Print all the violations from the database
    print_all_violations(DB_HOST, DB_USER, DB_PASSWORD, DB_NAME)
```
